Passo 2: Print do head com 10 linhas
Após carregar o arquivo, você pode visualizar as primeiras 10 linhas do dataframe usando o código abaixo:

```python
import pandas as pd

# Carregando o CSV
df = pd.read_csv('caminho/para/o/arquivo/carros.csv')

# Exibindo as primeiras 10 linhas
print(df.head(10))
```

Passo 3: Print das colunas e index do dataframe
Para visualizar as colunas e o índice do dataframe, você pode usar:

```python
print("Colunas do DataFrame:", df.columns)
print("Índice do DataFrame:", df.index)
```

Passo 4: Ordenação crescente pelo estado com maior quantidade de carros
Para ordenar o dataframe com base na quantidade de carros em cada estado de forma crescente, use:

```python
# Supondo que a coluna que contém a quantidade de carros seja 'Quantidade de Carros'
df_sorted = df.sort_values(by='Quantidade de Carros', ascending=True)
print(df_sorted)
```

Passo 5: Proporção de carro para a população de cada estado
Para calcular a proporção de carros por habitante em cada estado e adicionar uma nova coluna ao dataframe:

```python
# Supondo que as colunas sejam 'População' e 'Quantidade de Carros'
df['Proporção Carros/Habitante'] = df['Quantidade de Carros'] / df['População']
print(df)
```

Passo 6: Dataframe com Estado e proporção de carro por habitante
Para criar um novo dataframe com apenas o Estado e a proporção de carro por habitante:

```python
df_estado_proporcao = df[['Estado', 'Proporção Carros/Habitante']]
print(df_estado_proporcao)
```

Agregação de dados
Passo 7: Média de idade por região
Para calcular a média de idade por região:

```python
# Supondo que as colunas sejam 'Região' e 'Idade'
media_idade_regiao = df.groupby('Região')['Idade'].mean()
print(media_idade_regiao)
```

Passo 8: Porcentagem de pessoas com deficiência
Para calcular a porcentagem de pessoas com deficiência:

```python
# Supondo que a coluna 'Deficiencia' tenha valores binários, onde 1 = tem deficiência, 0 = não tem
porcentagem_deficiencia = df['Deficiencia'].mean() * 100
print(f"Porcentagem de pessoas com deficiência: {porcentagem_deficiencia:.2f}%")
```

Passo 9: Número de pessoas no estado do Piauí
Para contar quantas pessoas são do estado do Piauí:



```python
# Supondo que a coluna 'Estado' tenha os nomes dos estados
pessoas_piaui = df[df['Estado'] == 'Piauí'].shape[0]
print(f"Quantidade de pessoas no estado do Piauí: {pessoas_piaui}")
```
