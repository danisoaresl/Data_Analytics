Quando a tabela for gigante faz dessa forma.

```python
import pandas as pd
import numpy as np

from google.colab import drive
drive.mount('/content/drive')

# Certifique-se de que o nome do arquivo inclui '.gz'
file_path = '/content/drive/My Drive/game_medal_tally.csv.gz'

# Carregando o CSV comprimido em gzip
df = pd.read_csv(file_path, compression='gzip')

# Exibindo as primeiras linhas do DataFrame
print(df.head())

# Exibindo estatísticas descritivas
print(df.describe())

# Verificando valores nulos em cada coluna
null_counts = df.isnull().sum()
print("Contagem de valores nulos por coluna:")
print(null_counts)

# Verificando os nomes das colunas
print("Nomes das colunas no DataFrame:")
print(df.columns)

# Verificando os tipos de dados das colunas
print("Tipos de dados das colunas no DataFrame:")
print(df.dtypes)

# Verificando dados nulos na coluna 'medalhas' e tratando-os
if 'medalhas' in df.columns:
    print("Valores nulos na coluna 'medalhas':")
    print(df['medalhas'].isnull().sum())
    df['medalhas'] = df['medalhas'].fillna(0)  # Substitui valores nulos por 0

# Agrupando por país e somando as medalhas
try:
    if 'pais' in df.columns and 'medalhas' in df.columns:
        df_grouped = df.groupby('pais')['medalhas'].sum().reset_index()
        print("DataFrame agrupado por país e somando as medalhas:")
        print(df_grouped)
    else:
        print("As colunas 'pais' e/ou 'medalhas' não existem no DataFrame.")
except KeyError as e:
    print(f"Erro de chave: {e}. Verifique se as colunas 'pais' e 'medalhas' existem no DataFrame.")
except Exception as e:
    print(f"Erro inesperado: {e}")

```
